```python
import csv
import pandas as pd
import requests
from bs4 import BeautifulSoup as bs
def req(link):
    header= {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/101.0.4951.41 Safari/537.36 Edg/101.0.1210.32",
            "Accept-Encoding":"gzip, deflate", 
               "Accept":"text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8", 
               "DNT":"1",
               "Connection":"close", 
               "Upgrade-Insecure-Requests":"1"}
    url=link      
    response=requests.get(url, headers=header)
    return response

def scrap_details(item):
    
    #contact details
    details=item.find_all('div',{'class':'SERO-box mb15'})
    try:
        con=details[0].find_all('li')
        con1=[i.text.strip() for i in con]
        contact=" ".join(con1)
    
        #vice-chancellor details
        vice=details[1].find_all('li')
        vice_chan=[i.text.strip() for i in vice]
        vc=" ".join(vice_chan)  


        #registrar details
        reg=details[2].find_all('li')
        reg1=[i.text.strip() for i in reg]
        registrar=" ".join(reg1)
        
    except IndexError:
        contact='' 
        vc=''   
        registrar=''

    #state details
    adr=item.find_all('div',{'class':'col-md-4'})
    try:
        temp=adr[2].find('div',{'class':'box200'}).text.strip().split()
        state=" ".join(temp[:-2])

        #address details
        adr1=adr[1].find('div',{'class':'box200'}).text.strip()
        adr1=adr1.split()
        address=" ".join(adr1+temp[-2:])
    except IndexError:   
        state=''
        address=''
    

    

    result=(contact,vc,registrar,address,state)
    return result
    
    
#Main program
    
links=['https://www.ugc.ac.in/stateuniversitylist.aspx?id=9U80RXV93k09HNgOCkd8+w==&Unitype=iA7z/ds2To+MdcuCvitOpQ==',
'https://www.ugc.ac.in/stateuniversitylist.aspx?id=So1CNBLvrigKjpQTxHMrAQ==&Unitype=iA7z/ds2To+MdcuCvitOpQ==',
'https://www.ugc.ac.in/stateuniversitylist.aspx?id=Jc2qxh2PCqffjt5wPbleBA==&Unitype=iA7z/ds2To+MdcuCvitOpQ==',
'https://www.ugc.ac.in/stateuniversitylist.aspx?id=QX41wvEJ21KXlAW+tWNvlQ==&Unitype=iA7z/ds2To+MdcuCvitOpQ==',
'https://www.ugc.ac.in/stateuniversitylist.aspx?id=zxM7WatPi/jRM9v4021o7w==&Unitype=iA7z/ds2To+MdcuCvitOpQ==']

records=[]
for i in links:
    resp=req(i)
    soup=bs(resp.content, 'html.parser')
    univ_item=soup.find_all('div',{'class':'panel-body'})
    print(len(univ_item))
    for item in univ_item:
        record=scrap_details(item)
        records.append(record)
#         print(len(records))        
        
with open('upgrad-univ-scrap.csv','w',newline='',encoding='utf-8') as fp:
    writer=csv.writer(fp)
    writer.writerow(['Contact Information','VC Information','Registrar Information','Address','State'])
    writer.writerows(records)          
    
```
